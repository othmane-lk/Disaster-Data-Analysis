```python
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud
import re
from collections import Counter
from nltk.tokenize import word_tokenize
import nltk

# Download NLTK resources (run this only once)
nltk.download('punkt')


# Combine all comments into a single string
all_comments = ' '.join(df['Top Comment'])

# Tokenize the comments and count word frequencies
tokenized_words = word_tokenize(all_comments)
word_freq = Counter(tokenized_words)

data_analysis_terms = ['data', 'analysis', 'analytics', 'data analyst', 'data science', 'machine learning', 'statistics']

term_counts = {term: word_freq.get(term, 0) for term in data_analysis_terms}

# Displaying
print("Frequency of Relevant Terms:")
for term, count in term_counts.items():
    print(f"{term}: {count}")

plt.figure(figsize=(12, 6))
plt.bar(term_counts.keys(), term_counts.values(), color='skyblue')
plt.title('Frequency of Relevant Terms in Comments')
plt.xlabel('Terms')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.show()

# Visualize
wordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(term_counts)

plt.figure(figsize=(12, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud of Relevant Terms in Comments')
plt.show()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation

def perform_lda(comments, num_topics=5):
    vectorizer = CountVectorizer(max_df=0.95, min_df=2, stop_words='english')
    dtm = vectorizer.fit_transform(comments)

    lda = LatentDirichletAllocation(n_components=num_topics, random_state=42)
    lda.fit(dtm)

    return lda, vectorizer

lda_model, vectorizer = perform_lda(df['Top Comment'], num_topics=5)
feature_names = vectorizer.get_feature_names_out()

# Displaying
for topic_idx, topic in enumerate(lda_model.components_):
    top_words_idx = topic.argsort()[:-10 - 1:-1]
    top_words = [feature_names[i] for i in top_words_idx]
    print(f"Topic #{topic_idx + 1}: {', '.join(top_words)}")

# Visualiziing
plt.figure(figsize=(12, 8))
for topic_idx, topic in enumerate(lda_model.components_):
    plt.bar([f'Topic {topic_idx + 1}'], [sum(topic)], color='skyblue')

plt.title('Topic Distribution in Comments')
plt.xlabel('Topics')
plt.ylabel('Word Count')
plt.show()
```

```python
import re
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud
from textblob import TextBlob
from collections import Counter
from nltk.tokenize import word_tokenize
import nltk


nltk.download('punkt')
data = {'Top Comment': [
    "I have the skills required for this job.",
    "Strong Python skills required.",
    "Excellent data analysis skills required.",
    "Skills required: SQL, Python, and machine learning.",
    "The required skills are data visualization and statistics."
]}

df = pd.DataFrame(data)

# Extract words following the phrase 'skills required' in each comment
pattern = re.compile(r'\b(?:skills required)\s+(\w+)\b', re.IGNORECASE)
matches = df['Top Comment'].str.extract(pattern)
words_after_skills_required = [word.lower() for word_list in matches.values.flatten() if pd.notna(word_list) for word in word_list.split()]

word_counts = Counter(words_after_skills_required)

# Display the top 10 words and their frequencies
print("Top 10 Words Following 'Skills Required':")
print(word_counts.most_common(10))

sentiments = [TextBlob(comment).sentiment.polarity for comment in df['Top Comment']]
df['Sentiment'] = sentiments

plt.figure(figsize=(10, 6))
plt.hist(df['Sentiment'], bins=30, color='skyblue', edgecolor='black')
plt.title('Sentiment Distribution of "Skills Required" Sections')
plt.xlabel('Sentiment Polarity')
plt.ylabel('Number of Comments')
plt.show()

wordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(word_counts)

plt.figure(figsize=(12, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud of Words Following "Skills Required"')
plt.show()
```
